```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
heart_disease_dataset = pd.read_csv('Heart_Disease_Prediction.csv')
```

```python
heart_disease_dataset['Heart Disease'].value_counts().plot(kind = 'bar', color=['green', 'red'])

plt.title('Heart Disease Distribution')
plt.xlabel('Heart Disease (green = No Disease, red = Disease)')
plt.ylabel('Number of Patients')
plt.show()
```

```python
X = heart_disease_dataset.drop('Heart Disease', axis=1)
Y = heart_disease_dataset['Heart Disease']
```

```python
X.shape
```

```python
Y.shape
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=45)
```

```python
model = LogisticRegression(max_iter = 10000)
model.fit(X_train, Y_train)
```

```python
Y_train_pred = model.predict(X_train)
train_accuracy = accuracy_score(Y_train, Y_train_pred) 
```

```python
Y_test_pred = model.predict(X_test)
test_accuracy = accuracy_score(Y_test, Y_test_pred) 
```

```python
print("Training accuracy is:",train_accuracy)
print("Test accuracy is:",test_accuracy)
```
